```python
# BTVN Bài 1

import math
import copy
import random

X = "X"
O = "O"
EMPTY = " "

# ===== NHẬP KÍCH THƯỚC =====
while True:
    try:
        N = int(input("Nhập kích thước bàn cờ N x N (vd: 3, 5, 10): "))
        if N < 3:
            print("Phải >= 3!")
            continue
        break
    except ValueError:
        print("Vui lòng nhập số nguyên!")

def initial_board():
    return [[EMPTY]*N for _ in range(N)]

def print_board(board):
    num = 1
    width = len(str(N*N))

    for i in range(N):
        row = []
        for j in range(N):
            cell = str(num) if board[i][j] == EMPTY else board[i][j]
            row.append(cell.rjust(width))
            num += 1

        print(" | ".join(row))

        if i < N-1:
            print("-" * (N * (width + 3) - 3))

def number_to_pos(n):
    n -= 1
    return (n // N, n % N)

def pos_to_number(i, j):
    return i * N + j + 1

def winner(board):
    for i in range(N):
        if board[i][0] != EMPTY and all(board[i][j] == board[i][0] for j in range(N)):
            return board[i][0]

    for j in range(N):
        if board[0][j] != EMPTY and all(board[i][j] == board[0][j] for i in range(N)):
            return board[0][j]

    if board[0][0] != EMPTY and all(board[i][i] == board[0][0] for i in range(N)):
        return board[0][0]

    if board[0][N-1] != EMPTY and all(board[i][N-1-i] == board[0][N-1] for i in range(N)):
        return board[0][N-1]

    return None

def is_terminal(board):
    if winner(board):
        return True
    return all(EMPTY not in row for row in board)

def heuristic(board):
    w = winner(board)
    if w == X:
        return 100
    if w == O:
        return -100
    return 0

def actions(board):
    return [(i, j) for i in range(N) for j in range(N) if board[i][j] == EMPTY]

def result(board, move, player):
    new_board = copy.deepcopy(board)
    i, j = move
    new_board[i][j] = player
    return new_board

MAX_DEPTH = 3 if N >= 5 else 9

def minimax(board, depth, is_max):
    if is_terminal(board) or depth == MAX_DEPTH:
        return heuristic(board), None

    if is_max:
        best_val = -math.inf
        best_move = None
        for move in actions(board):
            val, _ = minimax(result(board, move, X), depth+1, False)
            if val > best_val:
                best_val = val
                best_move = move
        return best_val, best_move
    else:
        best_val = math.inf
        best_move = None
        for move in actions(board):
            val, _ = minimax(result(board, move, O), depth+1, True)
            if val < best_val:
                best_val = val
                best_move = move
        return best_val, best_move

board = initial_board()

user = input("Chọn X hoặc O: ").upper()
while user not in ["X", "O"]:
    user = input("Nhập lại (X/O): ").upper()

ai = O if user == X else X
turn = random.choice([X, O])

print("Người đi trước là:", turn)

while True:
    print()
    print_board(board)

    if is_terminal(board):
        w = winner(board)
        print("\nWinner:" if w else "\nHòa!", w if w else "")
        break

    if turn == user:
        try:
            move = int(input(f"Chọn ô (1-{N*N}): "))
            if move < 1 or move > N*N:
                print("Số không hợp lệ!")
                continue

            i, j = number_to_pos(move)

            if board[i][j] != EMPTY:
                print("Ô đã có người đánh!")
                continue

            board[i][j] = user
            turn = ai

        except ValueError:
            print("Nhập sai!")
            continue
    else:
        _, move = minimax(board, 0, ai == X)
        if move is None:
            break

        i, j = move
        print(f"\nAI đánh vào ô {pos_to_number(i, j)}")

        board = result(board, move, ai)
        turn = user
```
