```python
# BTVN Bài 3

import math
import copy
import random

X = "X"
O = "O"
EMPTY = " "

while True:
    try:
        N = int(input("Nhập kích thước bàn cờ N x N (vd: 3,5,10): "))
        if N < 3:
            print("Phải >= 3")
            continue
        break
    except:
        print("Nhập số nguyên!")

def initial_board():
    return [[EMPTY]*N for _ in range(N)]

def print_board(board):
    num = 1
    width = len(str(N*N))

    for i in range(N):
        row = []
        for j in range(N):
            cell = str(num) if board[i][j] == EMPTY else board[i][j]
            row.append(cell.rjust(width))
            num += 1
        print(" | ".join(row))

        if i < N-1:
            print("-" * (N * (width + 3) - 3))

def number_to_pos(n):
    n -= 1
    return (n // N, n % N)

def pos_to_number(i, j):
    return i * N + j + 1

def winner(board):
    for i in range(N):
        if board[i][0] != EMPTY and all(board[i][j] == board[i][0] for j in range(N)):
            return board[i][0]

    for j in range(N):
        if board[0][j] != EMPTY and all(board[i][j] == board[0][j] for i in range(N)):
            return board[0][j]

    if board[0][0] != EMPTY and all(board[i][i] == board[0][0] for i in range(N)):
        return board[0][0]

    if board[0][N-1] != EMPTY and all(board[i][N-1-i] == board[0][N-1] for i in range(N)):
        return board[0][N-1]

    return None

def is_terminal(board):
    if winner(board):
        return True
    return all(EMPTY not in row for row in board)

def heuristic(board):
    w = winner(board)
    if w == X:
        return 1000
    if w == O:
        return -1000
    return 0

def actions(board):
    return [(i, j) for i in range(N) for j in range(N) if board[i][j] == EMPTY]

def result(board, move, player):
    new_board = copy.deepcopy(board)
    i, j = move
    new_board[i][j] = player
    return new_board

MAX_DEPTH = 3 if N >= 5 else 9

def alphabeta(board, depth, alpha, beta, is_max):
    if is_terminal(board) or depth == MAX_DEPTH:
        return heuristic(board), None

    if is_max:
        best_val = -math.inf
        best_move = None

        for move in actions(board):
            val, _ = alphabeta(result(board, move, X), depth+1, alpha, beta, False)

            if val > best_val:
                best_val = val
                best_move = move

            alpha = max(alpha, best_val)

            if beta <= alpha:
                break

        return best_val, best_move

    else:
        best_val = math.inf
        best_move = None

        for move in actions(board):
            val, _ = alphabeta(result(board, move, O), depth+1, alpha, beta, True)

            if val < best_val:
                best_val = val
                best_move = move

            beta = min(beta, best_val)

            if beta <= alpha:
                break

        return best_val, best_move

board = initial_board()

user = input("Chọn X hoặc O: ").upper()
while user not in ["X", "O"]:
    user = input("Nhập lại (X/O): ").upper()

ai = O if user == X else X
turn = random.choice([X, O])

print("Người đi trước là:", turn)

while True:
    print()
    print_board(board)

    if is_terminal(board):
        w = winner(board)
        print("\nWinner:" if w else "\nHòa!", w if w else "")
        break

    if turn == user:
        try:
            move = int(input(f"Chọn ô (1-{N*N}): "))
            if move < 1 or move > N*N:
                print("Sai!")
                continue

            i, j = number_to_pos(move)

            if board[i][j] != EMPTY:
                print("Ô đã dùng!")
                continue

            board[i][j] = user
            turn = ai

        except:
            print("Lỗi nhập!")
            continue

    else:
        _, move = alphabeta(board, 0, -math.inf, math.inf, ai == X)

        if move is None:
            break

        i, j = move
        print(f"\nAI đánh vào ô {pos_to_number(i, j)}")

        board = result(board, move, ai)
        turn = user
```

```python
# BTVN Bài 5

import math
import copy

EMPTY = "."
WHITE = "W"
BLACK = "B"

board = [
    ["BK", ".", ".", "BP"],
    [".", ".", ".", "."],
    [".", ".", ".", "."],
    ["WP", ".", ".", "WK"]
]

def print_board(b):
    num = 1
    for i in range(4):
        row = []
        for j in range(4):
            cell = str(num) if b[i][j] == "." else b[i][j]
            row.append(cell.rjust(2))
            num += 1
        print(" | ".join(row))
        if i < 3:
            print("-----------------")
    print()

def number_to_pos(n):
    n -= 1
    return (n // 4, n % 4)

def get_moves(board, player):
    moves = []
    dirs = [(-1,0),(1,0),(0,-1),(0,1)]

    for i in range(4):
        for j in range(4):
            if board[i][j].startswith(player):
                for di, dj in dirs:
                    ni, nj = i+di, j+dj
                    if 0 <= ni < 4 and 0 <= nj < 4:
                        if board[ni][nj] == "." or not board[ni][nj].startswith(player):
                            moves.append(((i,j),(ni,nj)))
    return moves

def make_move(board, move):
    new_board = copy.deepcopy(board)
    (i,j),(ni,nj) = move
    new_board[ni][nj] = new_board[i][j]
    new_board[i][j] = "."
    return new_board

def evaluate(board):
    score = 0
    for row in board:
        for cell in row:
            if cell == "WK":
                score += 10
            elif cell == "WP":
                score += 1
            elif cell == "BK":
                score -= 10
            elif cell == "BP":
                score -= 1
    return score

def is_terminal(board):
    flat = sum(board, [])
    return "WK" not in flat or "BK" not in flat

def alphabeta(board, depth, alpha, beta, maximizing):
    if depth == 0 or is_terminal(board):
        return evaluate(board), None

    player = WHITE if maximizing else BLACK

    if maximizing:
        best_val = -math.inf
        best_move = None
        for move in get_moves(board, player):
            val, _ = alphabeta(make_move(board, move), depth-1, alpha, beta, False)
            if val > best_val:
                best_val = val
                best_move = move
            alpha = max(alpha, val)
            if beta <= alpha:
                break
        return best_val, best_move
    else:
        best_val = math.inf
        best_move = None
        for move in get_moves(board, player):
            val, _ = alphabeta(make_move(board, move), depth-1, alpha, beta, True)
            if val < best_val:
                best_val = val
                best_move = move
            beta = min(beta, val)
            if beta <= alpha:
                break
        return best_val, best_move

while True:
    print_board(board)

    if is_terminal(board):
        print("Game Over")
        break

    try:
        move = input("Nhập (từ ô -> đến ô): ")
        a, b = map(int, move.split())

        i, j = number_to_pos(a)
        ni, nj = number_to_pos(b)

        if board[i][j] == ".":
            print("Ô nguồn trống!")
            continue

        if not board[i][j].startswith(WHITE):
            print("Bạn chỉ được đi quân W!")
            continue

        if ((i,j),(ni,nj)) not in get_moves(board, WHITE):
            print("Nước đi không hợp lệ!")
            continue

        board = make_move(board, ((i,j),(ni,nj)))

    except:
        print("Nhập sai!")
        continue

    if is_terminal(board):
        print_board(board)
        print("Bạn thắng!")
        break

    _, ai_move = alphabeta(board, 3, -math.inf, math.inf, False)
    if ai_move:
        print("AI đi")
        board = make_move(board, ai_move)
```
