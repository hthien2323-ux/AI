```python
# BTVN Bài 2

import heapq

class DeliveryAStar:
    def __init__(self, graph):
        self.graph = graph

    def heuristic(self, current, remaining):
        return 0

    def a_star(self, start):
        n = len(self.graph)
        pq = []
        heapq.heappush(pq, (0, start, [start], 0))

        best_cost = float('inf')
        best_path = None

        while pq:
            f, current, path, cost = heapq.heappop(pq)

            if len(path) == n:
                total_cost = cost + self.graph[current][start]
                if total_cost < best_cost:
                    best_cost = total_cost
                    best_path = path + [start]
                continue

            for neighbor in self.graph:
                if neighbor not in path:
                    new_cost = cost + self.graph[current][neighbor]
                    h = self.heuristic(neighbor, None)
                    heapq.heappush(pq, (new_cost + h, neighbor, path + [neighbor], new_cost))

        print("Đường đi tối ưu:", best_path)
        print("Chi phí:", best_cost)
        return best_path, best_cost


graph = {
    'A': {'A':0,'B':10,'C':15,'D':20},
    'B': {'A':10,'B':0,'C':35,'D':25},
    'C': {'A':15,'B':35,'C':0,'D':30},
    'D': {'A':20,'B':25,'C':30,'D':0}
}

solver = DeliveryAStar(graph)
solver.a_star('A')
```
